```python
# Store filepath in a variable
Tennis_2014 = "Tennis_2014.csv"
Tennis_2015 = "Tennis_2015.csv"
Tennis_2016 = "Tennis_2016.csv"
```

```python
# Read our Data file with the pandas library
# May not be necesary -- ASK TRAVIS ABOUT THE ENCODING ISSUE ENCOUNTERED:

Tennis_2014 = pd.read_csv(Tennis_2014, encoding = "ISO-8859-1")
Tennis_2015 = pd.read_csv(Tennis_2015, encoding = "ISO-8859-1")
Tennis_2016 = pd.read_csv(Tennis_2016, encoding = "ISO-8859-1")
```

```python
# Drop all rows with missing information
Tennis_2014 = Tennis_2014.dropna(how='any')
```

```python
def missing_values_table(df):
        # Total missing values
        mis_val = Tennis_2014.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(Tennis_2014)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(Tennis_2014.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
```

```python
missing_values_table(Tennis_2014)
```
